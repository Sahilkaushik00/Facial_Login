```python
import pandas as pd
data=pd.read_csv("Faces.csv")
df1=pd.DataFrame(data)
df1
```

```python
from io import BytesIO
from PIL import Image
import cv2
import face_recognition
import requests
import webbrowser
import numpy as np


image_urls=df1['avatar']
face_encodings=[]

for url in image_urls:
    response=requests.get(url)
    if response.status_code==200:
        img=Image.open(BytesIO(response.content))
        image=cv2.cvtColor(np.array(img),cv2.COLOR_RGB2BGR)
        encoding=face_recognition.face_encodings(image)
        if encoding:
            face_encodings.append(encoding[0])
           
        else:
            value_to_remove=url
            df1=df1[df1['avatar']!=value_to_remove]
            #remove face if not encoded
df1
```

```python
import sys
video_capture = cv2.VideoCapture(0)

# Initialize a set to keep track of matched profiles
matched_profiles = set()


while True:
    _, frame = video_capture.read()
    small_frame = cv2.resize(frame, (0, 0), fx=0.25, fy=0.25)
    rgb_small_frame = small_frame[:, :, ::-1]
   
    face_locations = face_recognition.face_locations(rgb_small_frame)
    input_encodings = face_recognition.face_encodings(rgb_small_frame, face_locations)
    count =0
    for face_encoding in input_encodings:
        matches = face_recognition.compare_faces(face_encodings, face_encoding,tolerance=0.5)
        distances = face_recognition.face_distance(face_encodings, face_encoding)
        best_match_index = np.argmin(distances)

        if matches[best_match_index] and best_match_index not in matched_profiles:
            print("Access Granted")
            matched_profiles.add(best_match_index)  # Add the matched profile to the set
            count+=1
            sys.exit()
        else:
            print("Access Denied")
            sys.exit()
    cv2.imshow("Finding Profile...", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video_capture.release()
cv2.destroyAllWindows()
```
